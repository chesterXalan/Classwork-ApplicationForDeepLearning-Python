```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path
```

```python
dir = r'F:\School\Graduate School\class\Applications for Deep Learning\images'
path = [str(p) for p in Path(dir).glob('*.jpg')]
print(path)
```

```python
def highPassFilter(im):
    kernel = np.array([[-1, -1, -1], [-1, 8, -1], [-1, -1, -1]])
    im_res = cv2.filter2D(im, cv2.CV_32F, kernel)
    return im_res.astype('uint8')

def AMFilter(im):
    kernel = np.array([[1, 1, 1], [1, 1, 1], [1, 1, 1]])
    im_res = cv2.filter2D(im, cv2.CV_32F, kernel)
    im_res /= 9
    return im_res.astype('uint8')

def sharpFilter(im):
    kernel = np.array([[-1, -1, -1], [-1, 9, -1], [-1, -1, -1]])
    im_res = cv2.filter2D(im, cv2.CV_32F, kernel)
    return im_res.astype('uint8')

def sobelFilter(im):
    kernel = np.array([[0, -2, -2], [2, 0, -2], [2, 2, 0]])
    im_res = cv2.filter2D(im, cv2.CV_32F, kernel)
    return im_res.astype('uint8')
```

```python
im = cv2.cvtColor(cv2.imread(path[0]), cv2.COLOR_BGR2GRAY)
im1 = highPassFilter(im)
im2 = AMFilter(im)
im3 = sharpFilter(im)
im4 = sobelFilter(im)
```

```python
plt.figure(figsize=(15, 10))
plt.subplot(221); plt.imshow(im, 'gray')
plt.subplot(222); plt.imshow(im1, 'gray')
plt.subplot(223); plt.imshow(im2, 'gray')
plt.subplot(224); plt.imshow(im3, 'gray')
plt.show()
```
